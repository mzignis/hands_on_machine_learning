### Early Stopping

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```
m = 100
X = 3 * np.random.randn(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1) * 2
plt.scatter(X, y, alpha=0.75)
```

```
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)
```

```
from sklearn.model_selection import train_test_split

X_poly_train, X_poly_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2)
```

```
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(max_iter=1, warm_start=True, penalty=None, learning_rate='constant', eta0=0.0005)
```

```
from sklearn.metrics import mean_squared_error

min_val_error = float('inf')
best_epoch = None
best_model = None
interations = 1000

for epoch in range(interations):
    sgd_reg.fit(X_poly_train, y_train)
    
```
